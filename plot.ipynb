```r
ggplot(ld_data, aes(x = BP, y = R2)) +
  geom_point(color = "blue") +
  geom_smooth(color = "red", se = FALSE) +
  xlab("Distance (kb)") +
  ylab("LD (R^2)") +
  ggtitle("LD Decay Analysis - Chromosome NC_000009.12")
```
