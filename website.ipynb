```python
import streamlit as st
import pandas as pd

import subprocess
```

```python
df = pd.read_csv('out/sites.txt', delimiter='\t', header=None)
df = df[df[0].str.startswith('N')]
df['chr'] = df[0].str.split('_').str[-1].str.split('.').str[0].astype(int)
df.columns = ['chr', 'pos', 'Chr']
df['com'] = df.chr + df.pos.astype(str)
# df = df[df.chr<24]
df.head()
```

```python
ids = pd.read_csv('in/SNP_annotated_only_with_id.vcf', delimiter='\t', skiprows=767, usecols=range(3))
ids.columns = ['chr', 'pos', 'id_']
ids['com'] = ids.chr + ids.pos.astype(str)
ids.head()
```

```python
ids.shape, df.shape
```

```python
mer = pd.merge(df, ids, on =['chr', 'pos'], how='inner')
mer[['chr', 'pos', 'id_']].to_csv('./in/in_new.snp', sep='\t', index=False)
```

```python
chr = 3
chr = mer.chr[mer.Chr==chr].unique()[0]
start = 0
end = 30000000

command = f'LDBlockShow -InVCF in/SNP_annotated_only_with_id.vcf -OutPut  out_wbs/D_num   -Region  {chr}:{start}:{end} -OutPng -SeleVar 1 -NoShowLDist 1000000000 -SpeSNPName in/in_new.snp -OutPng -ShowNum'
command_ = f'LDBlockShow -InVCF in/SNP_annotated_only_with_id.vcf -OutPut  out_wbs/D   -Region  {chr}:{start}:{end} -OutPng -SeleVar 1 -NoShowLDist 1000000000 -SpeSNPName in/in_new.snp -OutPng'
command2 = f'LDBlockShow -InVCF in/SNP_annotated_only_with_id.vcf -OutPut  out_wbs/R_num   -Region  {chr}:{start}:{end} -OutPng -SeleVar 2 -NoShowLDist 1000000000 -SpeSNPName in/in_new.snp -OutPng -ShowNum'
command2_ = f'LDBlockShow -InVCF in/SNP_annotated_only_with_id.vcf -OutPut  out_wbs/R   -Region  {chr}:{start}:{end} -OutPng -SeleVar 2 -NoShowLDist 1000000000 -SpeSNPName in/in_new.snp -OutPng'
subprocess.run('rm out_wbs/*', shell=True)
subprocess.run(command, shell=True)
subprocess.run(command_, shell=True)
subprocess.run(command2, shell=True)
subprocess.run(command2_, shell=True)
```
